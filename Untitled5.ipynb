```python
# train_multivariate_lstm.py
"""
Multivariate LSTM to predict next-step OHLC (Open, High, Low, Close).
Adapted from user's original single-series script. See file reference. :contentReference[oaicite:1]{index=1}
"""

import os
import math
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# ---------- User settings ----------
CSV_PATH = "/content/EOD-AAPL_in.csv"      # input CSV (must have Open, High, Low, Close columns)
SEQ_LEN = 60                   # lookback timesteps
TRAIN_RATIO = 0.75
MODEL_OUT = "multivar_lstm_model"
SCALER_OUT = "multivar_scaler.save"
EPOCHS = 30
BATCH_SIZE = 32
RANDOM_SEED = 7
# -----------------------------------

np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

# ---------- 1. Load data ----------
if not os.path.exists(CSV_PATH):
    raise FileNotFoundError(f"Input CSV not found: {CSV_PATH}")

df = pd.read_csv(CSV_PATH)
# expected columns: Date, Open, High, Low, Close, ... ; original script used usecols [1,2,3,4]
# Try to find columns
required_cols = ['Open','High','Low','Close']
for c in required_cols:
    if c not in df.columns:
        raise ValueError(f"Required column '{c}' not found in CSV. Columns available: {df.columns.tolist()}")

# keep only OHLC and drop rows with missing values
df_ohlc = df[required_cols].dropna().reset_index(drop=True)
print("Loaded data shape (rows,cols):", df_ohlc.shape)

# ---------- 2. Scale features (fit scaler on training data only ideally) ----------
scaler = MinMaxScaler(feature_range=(0,1))
scaled_all = scaler.fit_transform(df_ohlc.values)   # shape (n_samples, 4)
joblib.dump(scaler, SCALER_OUT)
print("Scaler saved to:", SCALER_OUT)

# ---------- 3. Create sequences ----------
def create_multivariate_sequences(data, seq_len):
    """
    data: numpy array shape (n_samples, n_features)
    returns:
      X -> (n_sequences, seq_len, n_features)
      Y -> (n_sequences, n_features)   # next-step target (one-step ahead)
    """
    X, Y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i : i + seq_len, :])
        Y.append(data[i + seq_len, :])
    return np.array(X), np.array(Y)

X, Y = create_multivariate_sequences(scaled_all, SEQ_LEN)
print("Created sequences X.shape, Y.shape:", X.shape, Y.shape)

# ---------- 4. Train-test split (by time order) ----------
n_train = int(X.shape[0] * TRAIN_RATIO)
X_train, Y_train = X[:n_train], Y[:n_train]
X_test, Y_test = X[n_train:], Y[n_train:]
print("Train shape:", X_train.shape, Y_train.shape, "Test shape:", X_test.shape, Y_test.shape)

# ---------- 5. Build model ----------
n_features = X.shape[2]
model = Sequential([
    LSTM(64, input_shape=(SEQ_LEN, n_features), return_sequences=False),
    Dense(32, activation='relu'),
    Dense(n_features, activation='linear')   # output Open,High,Low,Close predicted next-step
])
model.compile(optimizer='adam', loss='mse')
model.summary()

# ---------- 6. Train ----------
history = model.fit(X_train, Y_train, validation_data=(X_test, Y_test),
                    epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=2)

# Save model (SavedModel format)
MODEL_OUT = "multivar_lstm_model.keras"
model.save(MODEL_OUT)
print("Model saved to:", MODEL_OUT)
# ---------- 7. Predict and inverse-transform ----------
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

# inverse-transform (scaler expects shape (n_samples, n_features))
pred_train_inv = scaler.inverse_transform(pred_train)
pred_test_inv = scaler.inverse_transform(pred_test)
Y_train_inv = scaler.inverse_transform(Y_train)
Y_test_inv = scaler.inverse_transform(Y_test)

# ---------- 8. Evaluation per-feature ----------
def rmse(a,b): return math.sqrt(mean_squared_error(a,b))
feature_names = required_cols
print("\nEvaluation (per feature):")
for i, fn in enumerate(feature_names):
    print(f"--- {fn} ---")
    tr_rmse = rmse(Y_train_inv[:,i], pred_train_inv[:,i])
    te_rmse = rmse(Y_test_inv[:,i], pred_test_inv[:,i])
    te_mape = mean_absolute_percentage_error(Y_test_inv[:,i], pred_test_inv[:,i]) * 100.0
    print(f" Train RMSE: {tr_rmse:.4f}")
    print(f" Test  RMSE: {te_rmse:.4f}")
    print(f" Test  MAPE: {te_mape:.2f}%\n")

# ---------- 9. Plot sample results (last N points of test) ----------
def plot_feature(idx, title, n_points=200):
    # choose last n_points from test
    n = len(Y_test_inv)
    start = max(0, n - n_points)
    x = np.arange(start, n)
    plt.figure(figsize=(8,3))
    plt.plot(x, Y_test_inv[start:, idx], label='Actual')
    plt.plot(x, pred_test_inv[start:, idx], label='Predicted', alpha=0.8)
    plt.title(f"{title} - Actual vs Predicted (test set last {n_points} points)")
    plt.legend()
    plt.tight_layout()

for i, fn in enumerate(feature_names):
    plot_feature(i, fn, n_points=200)
plt.show()

# ---------- 10. Save final predictions CSV (test subset) ----------
out_df = pd.DataFrame(scaler.inverse_transform(Y_test), columns=feature_names)
pred_df = pd.DataFrame(pred_test_inv, columns=[f"{c}_pred" for c in feature_names])
timestamps = np.arange(len(out_df))  # no original timestamps here; user can map to Date if needed
save_df = pd.concat([out_df.reset_index(drop=True), pred_df.reset_index(drop=True)], axis=1)
save_df.to_csv("multivar_test_predictions.csv", index=False)
print("Saved test predictions to multivar_test_predictions.csv")
```
